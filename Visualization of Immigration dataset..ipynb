```python
# Using a scatter plot,visualising the trend of total immigrantion to Canada (all countries combined) for the years 1980 - 2013.
df_tot=pd.DataFrame(df_can[years].sum(axis=0))
df_tot.index=map(int,df_tot.index)
df_tot.head()
df_tot.reset_index(inplace=True)
df_tot.rename(columns={'index':'years',0:'Total'},inplace=True)
df_tot.head()
```

```python
df_tot.plot(kind='scatter',x='years',y='Total',figsize=(10,6))
plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.grid()
plt.show()
```

```python
#Getting the best fit line for the above graph
x=df_tot['years']  # x-cordnate of data.
y=df_tot['Total']  # y-cordinate of data.
fit=np.polyfit(x,y,deg=1) #deg=1 linear 
print('Slope:',fit[0],', intercept:',fit[1])
```

```python
df_tot.plot(kind='scatter',x='years',y='Total',figsize=(10,6))
plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.plot(x,x*fit[0]+fit[1],color='red')           #Drawing the best fit line
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 200000))
plt.grid()
plt.show()
```

```python
# Creating a scatter plot of the total immigration from Denmark, Norway, and Sweden to Canada from 1980 to 2013?
df_countries=df_can.loc[['Denmark','Norway','Sweden'],years].sum(axis=0)
df_countries=pd.DataFrame(df_countries)
df_countries.index=map(int,df_countries.index)
df_countries.reset_index(inplace=True)
df_countries.rename(columns={'index':'Years',0:'Total'},inplace=True)
df_countries.head()
x=df_countries['Years']
y=df_countries['Total']
fit_new=np.polyfit(x,y,deg=1)
print('Slope:',fit_new[0],', intercept:',fit_new[1])
```

```python
df_countries.plot(kind='scatter',x='Years',y='Total',figsize=(10,6))
plt.title('Immigration from Denmark, Norway, and Sweden to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.plot(x,fit_new[0]*x+fit_new[1],color='red')
plt.grid()
plt.show()
```

```python
#Analysing Argentina great depression.
df_can_t=df_can[years].transpose()
df_can_t.index=map(int,df_can_t.index)
df_can_t.index.name='Years'
df_can_t.reset_index(inplace=True)
df_can_t.head()
```

```python
#Normalizing Brazil data
norm_brazil=((df_can_t['Brazil']-df_can_t['Brazil'].min())/(df_can_t['Brazil'].max()-df_can_t['Brazil'].min()))

norm_Arg=((df_can_t['Argentina']-df_can_t['Argentina'].min())/(df_can_t['Argentina'].max()-df_can_t['Argentina'].min()))
```

```python
ax0=df_can_t.plot(kind='scatter',
                 figsize=(10,6),
                 x='Years',
                 y='Brazil',
                s=norm_brazil * 2000 + 10,  # pass in weights 
                xlim=(1975, 2015))
ax1=df_can_t.plot(kind='scatter',
                 figsize=(10,6),
                 x='Years',
                 y='Brazil',
                color='red',
                  alpha=0.5,
                s=norm_Arg * 2000 + 10,  # pass in weights 
                xlim=(1975, 2015),
                 ax=ax0)
ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```
